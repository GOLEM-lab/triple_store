```python
for author,key in zip(df["Author"], df["Author_Key"]):
    string = """prefix golem: <https://golemlab.eu/graph/>
    WITH GRAPH <http://example.com> 
    DELETE {{ ?s golem:author "{}" .
    }} INSERT {{ ?s golem:author "{}" .
    }} WHERE {{ ?s golem:author "{}" .
    }} """.format(author,key, author)

sparql.setQuery(string)
results = sparql.query()
print("Done", author)
print(results.response.read())
```

```python
prefix_string = """prefix golem: <https://golemlab.eu/graph/>
    WITH GRAPH <http://example.com>"""

query_string = """DELETE {{ ?s golem:author "{}" .
    }} INSERT {{ ?s golem:author "{}" .
    }} WHERE {{ ?s golem:author "{}" .
    }}"""

query = prefix_string
count = 0
total_count = 0

while total_count <= 9981 : 
    while count < 50 :
        for author,key in zip(df["Author"], df["Author_Key"]):
            count+=1
            query += query_string.format(author, key, author)
            
    sparql.setQuery(query)
    results = sparql.query()
    print(results.response.read())
    
    total_count +=count 
    count = 0
```
